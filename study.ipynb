```python
import pandas as pd
import seaborn as sns
import seaborn.timeseries
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dateparse = lambda x: pd.datetime.strptime(x, '%Y-%m-%d')

global_chart = pd.read_csv('./sources/daily-2019-global.csv', parse_dates=['Date'], date_parser=dateparse)
mx_chart  = pd.read_csv('./sources/daily-2019-mx.csv', parse_dates=['Date'], date_parser=dateparse)
global_chart['Chart'] = 'Global'
mx_chart['Chart'] = 'Mexico'

resume = pd.concat([global_chart,mx_chart],keys=['Global','Mexico'])
```

```python
mx_unique = pd.read_csv('./sources/daily-2019-mx.csv')
mx_unique=mx_unique.rename(columns = {'Track Name':'track_name'})
mx_unique = mx_unique[mx_unique['Position'] == 1]
mx_unique = mx_unique.drop_duplicates('track_name')
mx_unique
```

```python
mx_unique = mx_unique[mx_unique['Artist'] == 'Billie Eilish']
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models.tools import HoverTool
from bokeh.models import BoxAnnotation
from bokeh.io import output_notebook

output_notebook()

p = figure(plot_width=400, plot_height=400)

p.title.text = "Valence - Arousal in Global Top 50"
p.circle(source=mx_unique, x='valence', y='energy')

green_box = BoxAnnotation(left=0.499, right=0.501, fill_color='black', fill_alpha=0.7)
p.add_layout(green_box)

green_box = BoxAnnotation(bottom=0.499, top=0.501, fill_color='black', fill_alpha=0.7)
p.add_layout(green_box)

hover=HoverTool()
hover.tooltips=[
    ('Track Name','@track_name'),
    ('Artist', '@Artist'),
    ('Valence','@valence'),
    ('Energy','@energy')
]
p.add_tools(hover)
show(p)
```

```python
global_emotion = pd.read_csv('./sources/daily-2019-global-unique-emotion.csv', parse_dates=['Date'], date_parser=dateparse)
global_emotion=global_emotion.rename(columns = {'Track Name':'track_name'})
global_emotion = global_emotion.drop_duplicates('track_name')
global_emotion
```

```python
from bokeh.models import TapTool 
colormap = {'happy': 'orange', 'excited': 'red', 'sad': 'blue', 'peacefull': 'green'}
colors = [colormap[x] for x in global_emotion['emotion']]
global_emotion['color'] = colors

p = figure(title = "Global Top 1 Tracks Valence-Arousal Analysis")
p.xaxis.axis_label = 'Valence'
p.yaxis.axis_label = 'Arousal(Energy)'

p.circle('valence', 'energy',source=global_emotion,legend_group='emotion',
         color='color', fill_alpha=0.2, size=10)

green_box = BoxAnnotation(left=0.499, right=0.501, fill_color='black', fill_alpha=0.7)
p.add_layout(green_box)

green_box = BoxAnnotation(bottom=0.499, top=0.501, fill_color='black', fill_alpha=0.7)
p.add_layout(green_box)

hover=HoverTool()
hover.tooltips=[
    ('Track Name','@track_name'),
    ('Artist', '@Artist'),
    ('Valence','@valence'),
    ('Energy','@energy')
]
p.add_tools(hover, TapTool())
p.legend.location = "bottom_right"

output_file("global_emotions.html", title="Global Chart Valence-Arousal Analysis")

show(p)
```

```python
global_chart = pd.read_csv('./sources/daily-2019-global.csv', parse_dates=['Date'], date_parser=dateparse)
top_1 = global_chart[global_chart['Position'] == 1]
top_1 = top_1.drop_duplicates('Track Id')
top_names = top_1['Track Id']
top_names = list(top_names)

global_top_tracks = global_chart.loc[global_chart['Track Id'].isin(top_names)]
global_top_tracks
```

```python
g = sns.relplot(x="Date", y="Position", kind="line", data=global_top_tracks, hue="Track Name")
g.set(ylim=(50, -1))
g._legend.remove()
g.fig.set_size_inches(25,6)
```

```python
global_top_tracks=global_top_tracks.rename(columns = {'Track Name':'track_name'})
senorita = global_top_tracks[global_top_tracks['track_name'] == 'Se√±orita']
ariana = global_top_tracks[global_top_tracks['track_name'] == 'thank u, next']
pm = global_top_tracks[global_top_tracks['track_name'] == 'Sunflower - Spider-Man: Into the Spider-Verse']
r_7 = global_top_tracks[global_top_tracks['track_name'] == '7 rings']
dy = global_top_tracks[global_top_tracks['track_name'] == 'Con Calma']
be = global_top_tracks[global_top_tracks['track_name'] == 'bad guy']
ty = global_top_tracks[global_top_tracks['track_name'] == 'ME! (feat. Brendon Urie of Panic! At The Disco)']
jb = global_top_tracks[global_top_tracks['track_name'] == "I Don't Care (with Justin Bieber)"]
cir = global_top_tracks[global_top_tracks['track_name'] == "Circles"]
```

```python
from bokeh.models import Range1d

p = figure(title = "Mexico's Top Hits", x_axis_type="datetime", plot_width=1000)

p.y_range=Range1d(50, 0)
p.xaxis.axis_label = 'Date'
p.yaxis.axis_label = 'Position'

p.line('Date','Position', source=senorita, line_width=2, line_color='#57B560', legend_label='Se√±orita')
p.line('Date', 'Position', source=ariana, line_width=2, line_color='#E57B9F', legend_label='thank u, next')
p.line('Date', 'Position', source=pm, line_width=2, line_color='#D9383B', legend_label='Sunflower - Spider-Man: Into the Spider-Verse')
p.line('Date', 'Position', source=r_7, line_width=2, line_color='#2E4AB2', legend_label='7 rings')
p.line('Date', 'Position', source=dy, line_width=2, line_color='#A13992', legend_label='Con Calma')
p.line('Date', 'Position', source=be, line_width=2, line_color='#F8E558', legend_label='bad guy')
p.line('Date', 'Position', source=ty, line_width=2, line_color='#5F9CEE', legend_label='ME! (feat. Brendon Urie of Panic! At The Disco)')
p.line('Date', 'Position', source=jb, line_width=2, line_color='#4C394F', legend_label="I Don't Care (with Justin Bieber)")
p.line('Date', 'Position', source=cir, line_width=2, line_color='#F7C974', legend_label='Circles')


hover=HoverTool(
    tooltips=[
        ('Track Name','@track_name'),
        ('Artist', '@Artist'),
        ('Position','@Position'),
        ('Date','@Date{%F}'),
        ('Streams', '@Streams')
    ],
    formatters={
        'Date': 'datetime'
    }
)

p.add_tools(hover, TapTool())
p.legend.location = "bottom_left"

output_file("mexico_top_1.html", title="Mexico's Top Hits")

show(p)
```
