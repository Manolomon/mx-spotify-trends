```python
mx_unique = pd.read_csv('./sources/daily-2019-mx-unique.csv')
mx_unique=mx_unique.rename(columns = {'Track Name':'track_name'})
```

```python
mx_unique[mx_unique['track_name'] == 'Amor A Primera Vista']
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models.tools import HoverTool
from bokeh.models import BoxAnnotation
from bokeh.io import output_notebook

output_notebook()

p = figure(plot_width=400, plot_height=400)

p.title.text = "Valence - Arousal in Global Top 50"
p.scatter(source=mx_unique, x='valence', y='energy')

green_box = BoxAnnotation(left=0.499, right=0.501, fill_color='black', fill_alpha=0.7)
p.add_layout(green_box)

green_box = BoxAnnotation(bottom=0.499, top=0.501, fill_color='black', fill_alpha=0.7)
p.add_layout(green_box)

hover=HoverTool()
hover.tooltips=[
    ('Track Name','@track_name'),
    ('Artist', '@Artist'),
    ('Valence','@valence'),
    ('Energy','@energy')
]
p.add_tools(hover)
show(p)
```
